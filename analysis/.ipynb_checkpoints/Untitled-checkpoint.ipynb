```python
import pandas as pd
import numpy as np
import plotly
import plotly.express as px
```

```python
first = pd.read_csv('Data.csv')
data = first[['FlightDate', 'HourRange', 'NonUsaPassengerAverageWait', 'UsaPassengerAverageWait']]
```

```python
data['FlightDate'] = pd.to_datetime(data['FlightDate'])
```

```python
data
```

```python
data['Year'] = data['FlightDate'].dt.year
data['Month'] = data['FlightDate'].dt.month
data['Day'] = data['FlightDate'].dt.dayofweek
```

## Day of week (integer, Monday=0 to Sunday=6)

```python
data
```

```python
data['hour'] = data['HourRange'].str.slice(0,2).astype(int)
```

```python
data['sin_hour'] = np.sin(2 * np.pi * data['hour'] / 24)
data['cos_hour'] = np.cos(2 * np.pi * data['hour'] / 24)
```

```python
data = data.drop(['FlightDate', 'HourRange'], axis = 1)
data
```

```python
temp2 = pd.DataFrame(data.groupby('hour')['NonUsaPassengerAverageWait'].mean())
fig = px.line(
    temp2, 
    x=temp2.index,
    y="NonUsaPassengerAverageWait", 
    title="Hour vs Non-USA Passenger Average Wait"
)

# Show the interactive plot
fig.show()
```

```python
temp3 = data.groupby('hour')['UsaPassengerAverageWait'].mean()
fig = px.line(
    temp3, 
    x=temp3.index,
    y="UsaPassengerAverageWait", 
    title="Hour vs USA Passenger Average Wait"
)

# Show the interactive plot
fig.show()
```

```python
usa_df = data.copy()
usa_df["citizenship"] = 1   # 1 = USA
usa_df["wait_time"] = usa_df["UsaPassengerAverageWait"]
usa_df
```

```python
nonusa_df = data.copy()
nonusa_df["citizenship"] = 0   # 0 = Non-USA
nonusa_df["wait_time"] = nonusa_df["NonUsaPassengerAverageWait"]
nonusa_df
```

```python
long_df = pd.concat([usa_df, nonusa_df], ignore_index=True)
long_df = long_df.drop(columns=["UsaPassengerAverageWait", "NonUsaPassengerAverageWait"])
long_df
```

## BASELINE

```python
features = [
    "citizenship",
    "Year",
    "Month",
    "Day",
    "sin_hour",
    "cos_hour"
]

X = long_df[features]
y = long_df["wait_time"]
```

```python
train_idx = long_df["Year"] < 2025
test_idx  = long_df["Year"] == 2025

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
print("Overall MAE:", mean_absolute_error(y_test, y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Overall MAE:", mean_absolute_error(y_test, y_pred_rf))
```

```python
from sklearn.metrics import mean_absolute_error

for group, name in [(1, "USA"), (0, "Non-USA")]:
    mask = X_test["citizenship"] == group
    mae = mean_absolute_error(y_test[mask], y_pred_rf[mask])
    print(f"{name} MAE: {mae:.2f}")
```

```python
baseline_pred = y_train.mean()
baseline_mae = mean_absolute_error(y_test, [baseline_pred] * len(y_test))
print("Mean baseline MAE:", baseline_mae)
```

```python
long_df["sin_day"] = np.sin(2 * np.pi * long_df["Day"] / 7)
long_df["cos_day"] = np.cos(2 * np.pi * long_df["Day"] / 7)
long_df
```

```python
long_df = long_df.sort_values(
    by=["Year", "Month", "Day", "hour", "citizenship"],
    ascending=[True, True, True, True, False]  # US (1) first
).reset_index(drop=True)
long_df
```

```python

long_df["lag_1"] = (
    long_df
    .groupby("citizenship")["wait_time"]
    .shift(1)
)

long_df["lag_3"] = (
    long_df
    .groupby("citizenship")["wait_time"]
    .shift(3)
)

long_df["rolling_3"] = (
    long_df
    .groupby("citizenship")["wait_time"]
    .rolling(3)
    .mean()
    .reset_index(level=0, drop=True)
)

long_df = long_df.dropna().reset_index(drop=True)
```

```python
features = [
    "citizenship",
    "Year", "Month", "Day",
    "sin_hour", "cos_hour",
    "sin_day", "cos_day"
]


X = long_df[features]
y = long_df["wait_time"]
```

```python
# Unique time blocks
time_blocks = long_df[["Year", "Month", "Day", "hour"]].drop_duplicates().reset_index(drop=True)

# 80 / 20 split by time
split_idx = int(len(time_blocks) * 0.8)

train_times = time_blocks.iloc[:split_idx]
test_times  = time_blocks.iloc[split_idx:]

# Map back to row indices
train_idx = long_df.merge(
    train_times, on=["Year", "Month", "Day", "hour"], how="inner"
).index

test_idx = long_df.merge(
    test_times, on=["Year", "Month", "Day", "hour"], how="inner"
).index

# Final splits
X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(
    objective="reg:squarederror",

    # --- Tree complexity ---
    max_depth=5,
    min_child_weight=5,
    gamma=0.5,

    # --- Boosting ---
    learning_rate=0.03,
    n_estimators=1500,

    # --- Regularization ---
    subsample=0.75,
    colsample_bytree=0.75,
    reg_lambda=2.0,
    reg_alpha=0.5,

    # --- Performance ---
    random_state=42,
    n_jobs=-1,
    tree_method="hist"
)
```

```python
feature_test = {
    "citizenship":1,
    "Year":2023, "Month":1, "Day":0,
    "sin_hour":1, "cos_hour":6.123234e-17,
    "sin_day":0, "cos_day":1
}
test = pd.DataFrame([feature_test])
model.predict(test)
```

```python
X.iloc[[0]]
```

```python
import xgboost
from xgboost import callback
model.fit(
    X_train,
    y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)
```

```python
pr = model.predict(X.iloc[[0]])
pr
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

y_pred = model.predict(X_test)

mae  = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"MAE : {mae:.2f}")
print(f"RMSE: {rmse:.2f}")

```

```python
test_df = long_df.iloc[test_idx].copy()
test_df["pred"] = y_pred

us_mae = mean_absolute_error(
    test_df.query("citizenship == 1")["wait_time"],
    test_df.query("citizenship == 1")["pred"]
)

non_us_mae = mean_absolute_error(
    test_df.query("citizenship == 0")["wait_time"],
    test_df.query("citizenship == 0")["pred"]
)

print("US MAE     :", us_mae)
print("Non-US MAE :", non_us_mae)
```

## FINAL TESTING

```python
test_df = long_df.iloc[test_idx].copy()
test_df["prediction"] = y_pred
test_df["residual"] = test_df["wait_time"] - test_df["prediction"]
test_df["abs_residual"] = np.abs(test_df["residual"])
```

```python
res_hour = (
    test_df
    .groupby(["hour", "citizenship"])["residual"]
    .mean()
    .reset_index()
)
```

```python
import matplotlib.pyplot as plt

for c, label in [(1, "US"), (0, "Non-US")]:
    subset = res_hour[res_hour["citizenship"] == c]
    plt.plot(subset["hour"], subset["residual"], label=label)

plt.axhline(0, linestyle="--")
plt.xlabel("Hour of Day")
plt.ylabel("Mean Residual (Actual - Predicted)")
plt.title("Residuals by Hour of Day")
plt.legend()
plt.show()
```
